```
from pycaret.utils import version
version()
```

```
from pycaret.datasets import get_data
dataSets = get_data('index')
```

```
jewelleryDataSet = get_data("jewellery")    # SN is 30
# This is unsupervised dataset.
# No target is defined.
```

```
from pycaret.clustering import *
kMeanClusteringParameters = setup(jewelleryDataSet)
print(kMeanClusteringParameters)
# Re-run the code if any error occur
```

```
# train kmeans model
kmeans = create_model('kmeans')
```

```
def apply_model(model_name):
    setup(data=jewelleryDataSet, verbose=False)

    print("With no preprocessing")

    print("For Cluster = 3")
    x = create_model(model_name, num_clusters=3)

    print("For Cluster = 4")
    x = create_model(model_name, num_clusters=4)


    print("For Cluster = 5")
    x = create_model(model_name, num_clusters=5)

    print("With Normalization")
    setup(data=jewelleryDataSet, normalize=True, normalize_method='zscore', verbose=False)

    print("For Cluster = 3")
    x = create_model(model_name, num_clusters=3)

    print("For Cluster = 4")
    x = create_model(model_name, num_clusters=4)

    print("For Cluster = 5")
    x = create_model(model_name, num_clusters=5)

    print("With Transformation")
    setup(data=jewelleryDataSet, transformation=True, transformation_method='yeo-johnson', verbose=False)

    print("For Cluster = 3")
    x = create_model(model_name, num_clusters=3)

    print("For Cluster = 4")
    x = create_model(model_name, num_clusters=4)

    print("For Cluster = 5")

    x = create_model(model_name, num_clusters=5)


    print("With Principle component analysis")
    setup(data=jewelleryDataSet, pca=True, pca_method='linear', verbose=False)

    print("For Cluster = 3")
    x = create_model(model_name, num_clusters=3)


    print("For Cluster = 4")

    x = create_model(model_name, num_clusters=4)


    print("For Cluster = 5")

    x = create_model(model_name, num_clusters=5)


    print("With Normalization + Transformation")
    setup(data=jewelleryDataSet, normalize=True, normalize_method='zscore',
          transformation=True, transformation_method='yeo-johnson', verbose=False)

    print("For Cluster = 3")

    x = create_model(model_name, num_clusters=3)


    print("For Cluster = 4")

    x = create_model(model_name, num_clusters=4)

    x = create_model(model_name)

    print("For Cluster = 5")

    x = create_model(model_name, num_clusters=5)


    print("With Normalization + Transformation + PCA")
    setup(data=jewelleryDataSet, normalize=True, normalize_method='zscore',
          transformation=True, transformation_method='yeo-johnson', pca=True, pca_method='linear', verbose=False)

    print("For Cluster = 3")

    x = create_model(model_name, num_clusters=3)


    print("For Cluster = 4")

    x = create_model(model_name, num_clusters=4)


    print("For Cluster = 5")

    x = create_model(model_name, num_clusters=5)


    print("Plotting elbow graph")
    evaluate_model(x)

    if(model_name != 'meanshift' ):
     plot_model(x, plot='elbow')

    if model_name != 'hclust' and model_name != 'meanshift' :  # Skip silhouette plot for models that do not support it
        print("Plotting silhouette plot")
        plot_model(x, plot='silhouette')


# Call the function with the desired clustering algorithm
apply_model('kmeans')  # or 'hclust', 'meanshift', etc.
```
